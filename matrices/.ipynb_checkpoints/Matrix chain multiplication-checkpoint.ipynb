```python
from numpy import arange, array, append
from numpy.random import randint
from time import time
from matplotlib.pyplot import legend, show, xlabel, ylabel, figure, loglog, title
from jupyterthemes import jtplot
from sys import maxsize
jtplot.style(theme = 'monokai')
```

```python
def best_parenthesization(s, i, j):
    l = []

    if(i == j):
        return l

    l += best_parenthesization(s, i, s[i][j])
    l += best_parenthesization(s, s[i][j] + 1, j)

    return l + [s[i][j]]
```

```python
def parenthesize(dims):
    n = len(dims) - 1
    m = [[0 for _ in range(n)] for _ in range(n)]
    s = [[0 for _ in range(n)] for _ in range(n - 1)]
    
    for o in range(n - 1):
        for i in range(n - o - 1):
            j = i + o + 1
            m[i][j] = maxsize
            
            for k in range(i, j):
                value = m[i][k] + m[k + 1][j] + dims[i] * dims[k + 1] * dims[j + 1]
            
                if value < m[i][j]:
                     m[i][j], s[i][j] = value, k

    return best_parenthesization(s, 0, n - 1)
```

```python
def parenthesize_memory_optimized(dims):
    n = len(dims) - 1
    m = [[0 for _ in range(n - i)] for i in range(n)]
    s = [[0 for _ in range(n - i - 1)] for i in range(n - 1)]
    
    for j in range(1, n):
        for i in range(n - j):
            m[i][j] = maxsize

            for k in range(j):
                value = m[i][k] + m[k + i + 1][j - k - 1] + dims[i] * dims[k + i + 1] * dims[j + i + 1]
                
                if value < m[i][j]:
                     m[i][j], s[i][j - 1] = value, k + i

    return best_draft(s, 0, n - 1)
```

```python
def complexity(func, args, *, nmax = 8):
    times = array([])
    sizes = 2 ** (1 + arange(nmax))

    for n in sizes:
        start = time()
        func(args(n))
        end = time()

        times = append(times, [end - start])

    return sizes, times
```

```python
size = 8
sizes, times = complexity(parenthesize, lambda n : randint(100, size = n), nmax = size)
x = array(arange(2, 2 ** (size))) 
```

```python
figure(figsize = [10, 10])
loglog(x, x ** 3 * times[-1] / x[-1] ** 3, 'y--', label = '$n^3$')
loglog(sizes, times, 'w:p', label = 'Dynamic programming solution') 
title('Time required with respect to the size of the input')
xlabel('$n$')
ylabel('$t$')
legend()
show()
```

```python
from typing import List

def findBreakingPoint(p: List[int], m: List[List[int]], s: List[List[int]],
                      i: int, j: int) -> None:
    m[i][j] = None
    s[i][j] = None

    for k in range(i, j):
        q = m[i][k] + m[k+1][j] + p[i]*p[k+1]*p[j+1]

        if m[i][j] is None or q < m[i][j]:
            m[i][j] = q
            s[i][j] = k


def unravel_best_parent(s: List[List[int]], i: int, j: int) -> List[int]:
    best = []

    if i < s[i][j] < j:
        best += unravel_best_parent(s, i, s[i][j])

    if i < s[i][j]+1 < j:
        best += unravel_best_parent(s, s[i][j]+1, j)

    return best + [s[i][j]]


def matrix_chain_mult(p: List[int]) -> List[int]:
    m = [[0 for i in range(len(p)-1)] for j in range(len(p)-1)]
    s = [[0 for i in range(len(p)-1)] for j in range(len(p)-1)]

    for diag in range(1, len(p)):
        for i in range(len(p)-diag-1):
            j = i + diag

            findBreakingPoint(p, m, s, i, j)

    return unravel_best_parent(s, 0, len(m)-1)
```
